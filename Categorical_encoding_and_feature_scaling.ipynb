```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import RobustScaler

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
le_gender = LabelEncoder()
df['Gender'] = le_gender.fit_transform(df['Gender'])
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# ----- Size → Ordered Label Encoding -----
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print(df)

# Apply RobustScaler to handle outliers in Fare
scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Justification:
# RobustScaler is used because it scales data using median and IQR,
# making it resistant to outliers like the high Fare value (450).

print(df)
```
